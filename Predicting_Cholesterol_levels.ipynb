**Quesiton 7: What variables predict the Cholesterol level?**

```python
import statsmodels.api as sm
import itertools

# Defining the x variables and y variables
x_variables = Cholesterol[['gender', 'height','weight','ap_hi', 'ap_lo', 'gluc','smoke','alco','cardio','Age (Years)' ]]
y_variables = Cholesterol['cholesterol_levels']

# Stepwise regression
def stepwise_regression(X, y):
    remaining_features = list(X.columns)
    best_score = -float('inf')
    best_model = None
    
    for k in range(1, len(X.columns) + 1):
        for subset in itertools.combinations(remaining_features, k):
            model = sm.OLS(y, X[list(subset)]).fit()
            score = model.rsquared_adj
            
            if score > best_score:
                best_score = score
                best_model = model
                best_subset = subset
                
        remaining_features = list(set(remaining_features) - set(best_subset))
        
    return best_model

model = stepwise_regression(x_variables, y_variables)

print(model.summary())

print("Regression Coefficients:")
for i, feature in enumerate(model.params.index):
    if i == 0:
        print(f"Intercept: {model.params[i]}")
    else:
        print(f"{feature}: {model.params[i]}")
        
print("\nP-values:")
print('BASED ON STEPWISE REGRESSION')
print(model.pvalues)

print(" ")
print("The 2 variables which are significant in predicting the cholesterol levels are 'weight' and 'Glucose levels'")        
```

**Question 8: Using the 2 variables from the OLS model, What is the regression line?**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Get the coefficients and intercept
coefficients = (model.coef_).round(4)
intercept = (model.intercept_).round(4)

# Print the equation of Yhat
equation = f"Yhat = {intercept} + {coefficients[0]}*weight + {coefficients[1]}*glucose"
print(equation)

# Printing the interpretation
print(" ")
print("""The intercept is 0.3916 which means that is the minimum cholesterol level or a constant for all the respondents.
The coefficient of weight is 0.0045 which means with an increase in 1 unit of weight, Cholesterol level is estimated to increase by 0.0045
The coefficient of Gluco level is 0.5248 which means with an increase in 1 level of Glucose, Cholesterol level is estimated to increase by 0.5248
      """)
```

**Question 9: Using the regression line equation, predict the cholesterol level for 2 individuals with weight [80,100] and gluco level [1, 0].**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming X is a 2D array or pandas DataFrame with the independent variables/features
glucose = Cholesterol['gluc']
weight = Cholesterol['weight']

# Reshape weight and glucose to 2D arrays with a single feature
X = np.column_stack((weight, glucose))

# Assuming y is a 1D array or pandas Series with the dependent variable/target
y = Cholesterol['cholesterol_levels']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict new values
new_data = np.array( [ [80, 1] , [100, 0] ] )
predicted_cholesterol = model.predict(new_data).round(2)

print("""The predicted cholesterol level for 2 individuals with weight [80,100] and gluco level [1,0] is : {}""".format(predicted_cholesterol))
```
